```python
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import Imputer
from sklearn.metrics import classification_report, accuracy_score
import math  
```

```python
def cross_validate_single_kernel(X, y, n_folds, c_vals, d_vals):
    
    error = 0
    
    skf=StratifiedKFold(n_splits=n_folds, shuffle=True)
    
    for train_index, test_index in skf.split(X, y):
        
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index].values.ravel(), y.iloc[test_index].values.ravel()
 
        ##normalization
        X_train_min, X_train_max = X_train.min(), X_train.max()
        
        X_train = (X_train - X_train_min)/(X_train_max-X_train_min)
        X_test  = (X_test  - X_train_min)/(X_train_max-X_train_min)

        X_train = X_train[X_train.columns[~(X_train_min==X_train_max)]]
        X_test  = X_test [X_test.columns[~(X_train_min==X_train_max)]]

        ##train/test
        svm = SVC(C=c_vals, kernel='poly', tol=1e-7, gamma = 'auto',  shrinking=False, degree = d_vals)
        
        svm.fit(X_train, y_train)

        pred = svm.predict(X_test)
        
        error = error + pow(accuracy_score(y_test, pred, normalize=True, sample_weight=None),2)
     
    print('c = ' + str(c_vals) + '  d = ' + str(d_vals))
     
    error = error/10 
    error = 1 - math.sqrt(error)   

    print(error)
    
    return error
```

```python
from sklearn.tree import DecisionTreeClassifier  
from sklearn import metrics 
from sklearn.model_selection import train_test_split

df = pd.read_csv('hw2_question3.csv')
df.head(5)
df.columns = ['Clump Thickness',  'Uniformity of Cell Size',  'Uniformity of Cell Shape', 
              'Marginal Adhesion',   'Single Epithelial Cell Size',   'Bare Nuclei', 
              'Bland Chromatin', 'Normal Nucleoli', 'Mitoses', 'test_Zebo_2_or_4']
df.head(5)
```

```python
df.groupby('test_Zebo_2_or_4').size() 
# as you can see from the result: the two classes are not equal
```

```python
feature_cols = ['Clump Thickness', 
              'Uniformity of Cell Size', 
              'Uniformity of Cell Shape', 
              'Marginal Adhesion', 
              'Single Epithelial Cell Size', 
              'Bare Nuclei', 
              'Bland Chromatin',
              'Normal Nucleoli',
              'Mitoses']
X = df[feature_cols]
Y = df['test_Zebo_2_or_4']
# Get the X and Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=1/3)
```

```python
# Decision tree by entropyentropy! 

clf = DecisionTreeClassifier(criterion="entropy") #  Decision Tree classifer object
clf = clf.fit(X_train,y_train)  # Train  

y_pred = clf.predict(X_test) #Predict 
 
print(metrics.accuracy_score(y_test, y_pred))

# Below we present with more visual elements: 

from IPython.display import Image  
import pydotplus
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO  

dot_data = StringIO()

export_graphviz(clf, out_file=dot_data, special_characters=True,feature_names = feature_cols,class_names=['benign','malignant'],filled=True, rounded=True)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  

graph.write_png('diabetes_entropy.png')

Image(graph.create_png())
```

```python
#  10-fold stratified CV while the maximum depth in the tree increases

from sklearn.model_selection import cross_val_score
import numpy as np
 
acc_entropy = []
acc_gini = []

for depth in range(1,20):
    clf_gini = DecisionTreeClassifier(max_depth=depth) # gini
    scores_gini = cross_val_score(clf_gini, X, y, cv=10)
    clf_entropy = DecisionTreeClassifier(max_depth=depth)
    scores_entropy = cross_val_score(clf_entropy, X, y, cv=10)
    acc_entropy.append(np.mean(scores_entropy))
    acc_gini.append(np.mean(scores_gini))
    
print(acc_entropy, acc_gini)
```

```python
from sklearn.ensemble import RandomForestRegressor

a = len(feature_list)
importances = [0]*a 
accuracies = 0;

# CV
for train_index, test_index in kf.split(features):
    
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    random_forrest = RandomForestRegressor(n_estimators = 2000, random_state = 38)
    
    random_forrest.fit(X_train, y_train);
    
    pred = random_forrest.predict(X_test)
    errors = abs(pred-y_test)
    
   # print('Error:', round(np.mean(errors), 2))
    
    mape = 100 * (errors / y_test)
    accuracy = 100 - np.mean(mape)
    accuracies += accuracy
    
  #  print('Accuracy:', round(accuracy, 2))
    
    cur_importances = list(random_forrest.feature_importances_)
    
    for k in range(0, len(cur_importances)):
        importances[k] += float(cur_importances[k])
    
for i in range(0,len(importances)):
    importances[i] /= 10.0
    
feature_importances = [ (feature, round(importance,2)) for feature, importance in zip(feature_list, importances)]

feature_importances = sorted(feature_importances, key = lambda x : x[1], reverse=True)

print(feature_importances)

accuracies /= 10.0

# print('average accuracy:', accuracies)
```
