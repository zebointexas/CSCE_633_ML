### Q4(a) 
Do SVD: First we need to read the gif data and transform it into a one-dimension array, then we normalize the data and proceed the svd.

```python
#importing libraries
import numpy as np
from PIL import Image
import os
from numpy import linalg as LA
import matplotlib.pyplot as plt
from sklearn import preprocessing
scalar = preprocessing.StandardScaler() #Standardize features by removing the mean and scaling to unit variance
```

```python
#stacking input images to a single matrix
directory = os.getcwd()
empty_arr = np.zeros(77760)

for filename in os.listdir(directory):
#     print('f',filename)
    if not filename.endswith('.ipynb_checkpoints'):
        temp_image = Image.open(filename)
        print(temp_image.size)
        image_size = temp_image.resize((320,243))
        image_arr = np.array(image_size) # Convert the image format into numpy array
        flatten_arr = image_arr.flatten() #flatten the matrix to one dimension array
        empty_arr = np.vstack((empty_arr,flatten_arr)) #replace empty_arr with flatten_arr
    
inp_arr = empty_arr[1:]
mean = np.mean(inp_arr,axis = 0) #get the mean value of input array

for i in inp_arr:
    i -= mean  #set the mean of array to 0.
    
inp_arr = scalar.fit_transform(inp_arr) #normalize the data
```

```python
#calculating singular value decomposition
U,eig_val,V = LA.svd(inp_arr.transpose())
```

```python
print(U.shape)
```

```python
#compute eigen_value and eigen vector
x = np.diag(eig_val)
eig_val,eig_vec = np.linalg.eig(x)
print('eig_value: ',eig_val)
print('eig_vector: ',eig_vec)
```

### Q4(2)
Plot eigen value and capture 50% Energy.

```python
plt.plot(range(len(eig_val)),eig_val)
plt.title('Display Eigen value') 
plt.show()
```

```python
energy = 0
count = 0
print('sum energy: ', sum(eig_val))
for item in eig_val:
    energy += item
    count += 1
    print('energy: ',energy)
    if energy/sum(eig_val)>=0.5:
        print('number of components needed: ', count)
        break
```

### Q(d)
Visualize the reconstructed images. From the construction figures below, we find that the more components used, the better reconstruction result.

```python
nComp = [1,10,20,30,40,50]

for item in nComp:
    Z = np.dot(inp_arr[0],U[:,:item]) #PCA results, inp_arr[1] is the index of person.
    # print(Z.shape)
    recon = np.dot(Z,U[:,:item].T) #do reconstruction
    # print(recon.shape)
    plt.imshow(recon.reshape((160,160)),cmap='gray')
    plt.show()
```
